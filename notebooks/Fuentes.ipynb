### Music Song

- **instance_id:** Un identificador único para cada instancia en el conjunto de datos, representado como un número de punto flotante. Hay un total de 50,000 instancias.
- **artist_name:** El nombre del artista que creó la canción, representado como una cadena (objeto).
- **track_name:** El nombre de la canción, representado como una cadena (objeto).
- **popularity:** Una medida de la popularidad de la canción, representada como un número de punto flotante entre 0 y 100. Valores más altos indican mayor popularidad.
- **acousticness:** Una medida de la acústica de la canción, representada como un número de punto flotante entre 0 y 1. Valores más altos indican mayor acústica.
- **danceability:** Una medida de la capacidad de baile de la canción, representada como un número de punto flotante entre 0 y 1. Valores más altos indican mayor capacidad de baile.
- **duration_ms:** La duración de la canción en milisegundos, representada como un número de punto flotante.
- **energy:** Una medida de la energía de la canción, representada como un número de punto flotante entre 0 y 1. Valores más altos indican mayor energía.
- **instrumentalness:** Una medida de la instrumentalidad de la canción, representada como un número de punto flotante entre 0 y 1. Valores más altos indican mayor instrumentalidad.
- **key:** La tonalidad en la que se interpreta la canción, representada como una cadena (objeto). Hay 12 valores posibles: 'C', 'C#', 'D', 'D#', 'E', 'F', 'F#', 'G', 'G#', 'A', 'A#' y 'B'.
- **liveness:** Una medida de la vivacidad de la canción, representada como un número de punto flotante entre 0 y 1. Valores más altos indican mayor vivacidad.
- **loudness:** Una medida del volumen de la canción, representada como un número de punto flotante en decibelios (dB). Valores más altos indican mayor volumen.
- **mode:** El modo en el que se interpreta la canción, representado como una cadena (objeto). Hay dos valores posibles: 'mayor' y 'menor'.
- **speechiness:** Una medida de la cantidad de habla en la canción, representada como un número de punto flotante entre 0 y 1. Valores más altos indican mayor cantidad de habla.
- **tempo:** El tempo de la canción en pulsos por minuto (BPM), representado como una cadena (objeto).
- **obtained_date:** La fecha en la que se obtuvo la información de la canción, representada como una cadena (objeto).
- **valence:** Una medida de la valencia (positividad) de la canción, representada como un número de punto flotante entre 0 y 1. Valores más altos indican mayor positividad.
- **music_genre:** El género de la canción, representado como una cadena (objeto). Hay varios géneros posibles en el conjunto de datos.
